```python
import tensorflow as tf
from tensorflow.keras import datasets, models, layers
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
(train_images,train_labels),(test_images,test_labels) = datasets.mnist.load_data()
```

```python
train_images = train_images.reshape(60000,28,28,1).astype('float32') / 255.0
test_images = test_images.reshape(10000,28,28,1).astype('float32') / 255.0
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation = 'relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
    ])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(train_images, train_labels, epochs=20)
```

```python
loss, acc = model.evaluate(test_images, test_labels)
```

```python
print("accuracy:",acc)
accu = acc*100
print(accu)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, transform

correct = 0
total = 0

for digit in range(1, 10):

    folder = f"/content/Handwritten Digits/{digit} image"

    for image_name in os.listdir(folder):

        image_path = os.path.join(folder, image_name)


        img = io.imread(image_path)


        if len(img.shape) == 3 and img.shape[2] == 4:
            img = color.rgba2rgb(img)

        if len(img.shape) == 3:
            img = color.rgb2gray(img)


        img = 1 - img


        rows = np.any(img > 0.2, axis=1)
        cols = np.any(img > 0.2, axis=0)

        if rows.any() and cols.any():

            y1, y2 = np.where(rows)[0][[0, -1]]
            x1, x2 = np.where(cols)[0][[0, -1]]

            img = img[y1:y2+1, x1:x2+1]


        h, w = img.shape

        size = max(h, w)

        square = np.zeros((size, size), dtype=np.float32)

        y_offset = (size - h) // 2
        x_offset = (size - w) // 2

        square[y_offset:y_offset+h,
               x_offset:x_offset+w] = img


        img = transform.resize(square,
                               (28,28),
                               anti_aliasing=True)

        img = img.astype("float32")


        plt.imshow(img, cmap='gray')
        plt.title(f"Actual: {digit}")
        plt.axis('off')
        plt.show()


        img = img.reshape(1,28,28,1)


        prediction = model.predict(img, verbose=0)

        predicted_digit = np.argmax(prediction)

        print(f"{image_name:15}  Actual:{digit}  Predicted:{predicted_digit}")

        if predicted_digit == digit:
            correct += 1

        total += 1

accuracy = (correct / total) * 100

print("\n==============================")
print("Total Images :", total)
print("Correct :", correct)
print("Wrong :", total - correct)
print(f"Accuracy : {accuracy:.2f}%")
```
